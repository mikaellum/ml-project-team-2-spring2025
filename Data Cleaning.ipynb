```python
import pandas as pd
import numpy as np
import librosa
import matplotlib.pyplot as plt
# import ast
# import re
# import json
```

```python
df = pd.read_pickle('../Database/music_database.pkl')

df.head()
```

```python
#load genre mapping CSV
genre_map_df = pd.read_csv("Genre_Mapping.csv")
print(genre_map_df.head())
```

```python
#replace with actual column names
mapping_dict = dict(zip(genre_map_df['OriginalLabel'], genre_map_df['MainGenre']))

#create a new column for mapped genre
df['main_genre'] = df['genre'].map(mapping_dict)
```

```python
df.head() # check new main_genre column
```

```python
#replace original genre with mapped version
df['genre'] = df['main_genre']
df = df.drop(columns=['main_genre'])

df.head() # confirm the result
```

```python
target_genres = [
    "Ambient/Experimental/Noise",
    "Children/Holiday/Novelty",
    "Classical/Cinematic/Orchestral",
    "Electronic/Dance",
    "Folk/Acoustic/Singer-Songwriter",
    "Hip-Hop/Rap",
    "Jazz/Blues/Soul",
    "Metal",
    "Other/Undefined",
    "Pop/Synth-Pop/Indie-Pop",
    "Reggae/Dub",
    "Rock/Alternative/Punk",
    "World/International/Latin"
]

df = df[df['genre'].isin(target_genres)] # remove everything that is not in target genre
print("Remaining unique genres:", df['genre'].nunique())
print("Final genre classes:", sorted(df['genre'].unique()))
```

```python
print(df.columns) #check new shape
print(df.head()) 
print(df.shape)
```

```python
#display unique values in year column to inspect inconsistent formats
total_entries = len(df)


#extract 4-digit year using regex
df['year_extracted'] = df['year'].astype(str).str.extract(r'(\d{4})')

#count how many rows have no 4-digit year
invalid_year_count = df['year_extracted'].isna().sum()

print(f"Total entries: {total_entries}")
print(f"Entries with invalid or missing year: {invalid_year_count}")
```

```python
#drop rows where extracted year is missing
df = df.dropna(subset=['year_extracted'])

#convert extracted year to integer
df['year'] = df['year_extracted'].astype(int)

df.drop(columns=['year_extracted'], inplace=True) #drop the temporary column


print(f"\nData shape after removal: {df.shape}")
print(df['year'].unique()) # check if year numbers are correct
```

```python
#find duplicate titles
duplicate_titles = df[df.duplicated(subset=['title'], keep=False)]

#display them
print(duplicate_titles.sort_values(by='title'))

#count how many duplicated titles are there
num_duplicates = df.duplicated(subset=['title']).sum()
print(f"Number of duplicated titles: {num_duplicates}")
```

```python
#remove duplicates, keep the first occurrence
df = df.drop_duplicates(subset=['title'], keep='first')

#confirm
print(f"Data shape after removing duplicates: {df.shape}")
```

```python
#drop Ch0 and Ch1 temporarily just for duplication check
meta_columns = df.drop(columns=['Ch0', 'Ch1'], errors='ignore')

##check for duplicates in metadata
num_full_duplicates = meta_columns.duplicated().sum()
print(f"Number of fully duplicated rows (excluding Ch0/Ch1): {num_full_duplicates}")
```

```python
#identify columns where all values are empty
meta_df = df.drop(columns=['Ch0', 'Ch1'], errors='ignore')
empty_cols = meta_df.columns[meta_df.isna().all()]
print(f"Columns with all values missing: {list(empty_cols)}")
```

```python
##sr=22050 is because 22050 Hz is the standard default in audio processing libraries like librosa, it represents number of samples per second
## MFCC represents the short-term power spectrum of sound, it sets to 13 values because 13 is commonly used in music/audio analysis
##chrome represents the intensity of each of the 12 pitch classes, therefore it sets to 12.
##RMS Energy represents the average loudness or energy of the audio. Useful for distinguishing dynamic levels. 

def extract_audio_features(row, sr=22050, n_mfcc=13):
    
    ch0, ch1 = row['Ch0'], row['Ch1']

    #this is to ensure both channels are converted to 1-D float32 arrays
    ch0 = np.asarray(ch0, dtype=np.float32).ravel()
    ch1 = np.asarray(ch1, dtype=np.float32).ravel()

    #if the two channels are of different lengths, crop both to the shorter length
    min_len = min(len(ch0), len(ch1))
    if min_len == 0:
        raise ValueError("Empty audio")
    ch0, ch1 = ch0[:min_len], ch1[:min_len]

    #convert stereo to mono by averaging the channels
    mono = (ch0 + ch1) / 2.0  

    #extract audio features: MFCC, Chroma, RMS Energy
    mfcc   = librosa.feature.mfcc(y=mono, sr=sr, n_mfcc=n_mfcc).mean(axis=1)
    chroma = librosa.feature.chroma_stft(y=mono, sr=sr).mean(axis=1)
    rms    = librosa.feature.rms(y=mono).mean()

    return np.concatenate([mfcc, chroma, [rms]]) #combine all features into one vector
```

```python
feature_list = [] ##will hold the extracted feature vectors
bad_indices   = [] #track rows that failed during feature extraction

for idx, row in df.iterrows(): #iterate over each row of the dataframe to process audio data
    try:
        feats = extract_audio_features(row) #extract features from Ch0 and Ch1
        feature_list.append(feats)
    except Exception as e:
        bad_indices.append(idx) #log the row index if extraction fails

print(f"Successful: {len(feature_list)}  |  Skipped: {len(bad_indices)}")
```

```python
#define column names for MFCCs, Chroma features, and RMS energy
feature_names = [f"mfcc_{i+1}" for i in range(13)] + \
                [f"chroma_{i+1}" for i in range(12)] + ["rms_energy"]

feature_df = pd.DataFrame(feature_list, columns=feature_names, index=df.drop(bad_indices).index) #create a dataframe from the extracted feature vectors
```

```python
#concatenate the features with the corresponding genre label (matching by index)
df_ml = pd.concat([feature_df, df.loc[feature_df.index, 'genre'].reset_index(drop=True)], axis=1)

print(df_ml.head())
```
